Create a scatter plot to represent the data. What can you conclude from the scatterplot? Make sure you have the appropriate labels for the x and y axis.

```python
temps = np.array([57.56, 61.52, 53.42, 59.36, 65.3, 71.78, 66.92, 77.18, 74.12, 64.58])
sales = np.array([215, 325, 185, 332, 406, 522, 412, 614, 544, 421])

df = pd.DataFrame({'Temperature': temps, 'Total Revenue': sales})

sns.scatterplot(data=df, x='Temperature', y='Total Revenue')
plt.title('Effect of Temperature on Total Revenue')
plt.xlabel('Temperature (degrees F)')
plt.ylabel('Total Revenue ($)')
plt.show()
```

There seems to be a pretty strong linear relationship between the outside temperature and this ice cream shop's total sales. As temperature increases, so does total revenue, in a linear fashion.

Create a line graph that displays the total profit across all months. Make sure you have the appropriate labels for the x and y axis. Optional: Split the line graph by ice cream flavor to compare sales by flavors over time.


```python
hist_sales_df = pd.read_csv('./Monthly Ice Cream Shop Sales by Flavor.csv')
hist_sales_df.head()
```

```python
sns.lineplot(data=hist_sales_df, x='Month', y='Total Profit')
plt.title("Total Profit by Month")
plt.show()
```

```python
hist_sale_flavors_only_df = hist_sales_df[['Vanilla','Chocolate', 'Strawberry', 'Neapolitan', 'Mango', 'Lime']]
display(hist_sale_flavors_only_df)

sns.lineplot(data=hist_sale_flavors_only_df)
plt.title("Total Sales Volume by Month per Flavor")
plt.xlabel('Month')
plt.ylabel('Number of Units')
plt.show()
```

Create a bar chart that displays the total number of units sold per month. Make sure you have the appropriate labels for the x and y axis.

```python
sns.barplot(data=hist_sales_df, x='Month', y='Total Units')
plt.title('Total Units Sold per Month')
plt.ylabel('Number of Units Sold')
plt.show()
```
